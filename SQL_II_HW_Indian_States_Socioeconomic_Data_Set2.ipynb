```python
#1. List all states in the dataset, sorted alphabetically.
pd.read_sql_query("Select state FROM state_stats Order By state asc",conn)
```

```python
#2. Show the states and regions where the GDP is below 5000 crore.
pd.read_sql_query("Select state, region, gdp_cr as gdp FROM state_stats WHERE gdp_cr < 5000",conn)
```

```python
#3. Which state has the highest literacy rate?
pd.read_sql_query("Select state, literacy_rate FROM state_stats Order By literacy_rate desc Limit 1",conn)
```

```python
#4. Show the 3 largest states by area.
pd.read_sql_query("Select state, area_km2 as area From  state_stats Order By area_km2 desc Limit 3 ",conn)
```

```python
#5. Which regions have at least one state with urbanization above 60%? (think: use DISTINCT region with WHERE)
pd.read_sql_query("Select  region, state, urbanization_pct From state_stats Where urbanization_pct >60 Group by  region",conn)
```

```python
#6. Show all states where population is less than the average population of all states.
pd.read_sql_query("Select state, population_lakh From state_stats Where Population_lakh < (Select avg(population_lakh) From state_stats)",conn)
```

```python
#7. For each region, what is the highest literacy rate?
pd.read_sql_query("Select region, max(literacy_rate) From state_stats Group By region",conn)
```

```python
#8. List the states that have a literacy rate equal to the average literacy rate of all states (if any).
pd.read_sql_query("Select state, literacy_rate From state_stats Where literacy_rate = (Select avg(literacy_rate) From state_stats)",conn)
```

```python
pd.read_sql_query("Select * from state_stats limit 1",conn)
```

```python
#9. What’s the total GDP of all southern states?
pd.read_sql_query("Select region, sum(gdp_cr) From state_stats Where region = 'South'",conn)
```

```python
#10. Show all states whose name starts with the letter ‘K’.
pd.read_sql_query("Select state From state_stats Where state like 'K%'",conn)
```

```python
#11. Which state(s) have both the minimum and maximum population in the dataset?
pd.read_sql_query('''SELECT state, population_lakh
                      FROM state_stats
                      WHERE population_lakh = (Select min(population_lakh) FROM state_stats)
                      OR
                      population_lakh = (Select max(population_lakh) From state_stats)
                     ''',conn)
```

```python
#12. For each region, how many states have a GDP above 1 lakh crore? (GROUP BY + WHERE)
pd.read_sql_query("SELECT region, count(state), gdp_cr FROM state_stats WHERE gdp_cr > 100000  GROUP BY region",conn)
```

```python
#13. Which states have a literacy rate between 75 and 80 (inclusive)?
pd.read_sql_query("SELECT state, literacy_rate FROM state_stats WHERE literacy_rate BETWEEN 75 and 80",conn)
```

```python
#14. Show all states whose name contains ‘Pradesh’.
pd.read_sql_query("SELECT state FROM state_stats WHERE state like '%Pradesh'",conn)
```

```python
#15. List the regions, sorted by their average urbanization percentage (highest first).
pd.read_sql_query("SELECT region, avg(urbanization_pct) as avg_urbanization FROM state_stats GROUP BY region ORDER BY avg_urbanization desc",conn)
```

```python
#16. Which state(s) have a population closest to the median population of all states? (trick! requires creative use of ORDER BY & LIMIT)
pd.read_sql_query("SELECT state, population_lakh FROM state_stats ORDER BY population_lakh desc Limit 1 OFFSET 16  ",conn)
```

```python
#17. List all states with a GDP per sq km (gdp_cr / area_km2) above 1. (calculated field!)
pd.read_sql_query("SELECT state, gdp_cr, area_km2 FROM state_stats WHERE gdp_cr/area_km2 > 1",conn)
```

```python
# 1. List all the states in the dataset.
pd.read_sql_query("SELECT state FROM state_stats",conn)
```

```python
#2. Which states have a literacy rate above 85?
pd.read_sql_query("SELECT state, literacy_rate FROM state_stats WHERE literacy_rate>85",conn)
```

```python
#3. Show all details of states from the ‘South’ region.
pd.read_sql_query("SELECT * FROM state_stats WHERE region = 'South'",conn)
```

```python
#4. List the unique regions present in the dataset.
pd.read_sql_query("SELECT DISTINCT(region) as Unique_Regions FROM state_stats",conn)
```

```python
#5. Which states have a population greater than 1000 lakh?
#pd.read_sql_query("SELECT * FROM state_stats limit 1",conn)
pd.read_sql_query("SELECT state, population_lakh as Population FROM state_stats WHERE population_lakh>1000",conn)
```

```python
#6. What is the average literacy rate across all states?
pd.read_sql_query("SELECT round(AVG(literacy_rate),2) as Avg_literacy_rate FROM state_stats",conn)
```

```python
#7. Show all states with urbanization percentage above 50, sorted by urbanization percentage descending.
pd.read_sql_query("SELECT state, urbanization_pct FROM state_stats WHERE urbanization_pct>50 ORDER BY urbanization_pct desc",conn)
```

```python
#8. How many states are there in each region?
pd.read_sql_query("SELECT region, count(state) as no_of_states FROM state_stats GROUP BY region",conn)
```

```python
#9. List the top 5 states by GDP (gdp_cr), sorted highest to lowest.
pd.read_sql_query("SELECT state, gdp_cr FROM state_stats ORDER BY gdp_cr desc limit 5",conn)
```

```python
#10. Which state has the smallest area?
pd.read_sql_query("SELECT state, min(area_km2) as area FROM state_stats ",conn)
```

```python
#11. For each region, show the average GDP.
pd.read_sql_query("SELECT region, round(avg(gdp_cr),2) as gdp FROM state_stats GROUP BY region",conn)
```

```python
#12. List all states where population is below 100 lakh and literacy rate is above 80.
pd.read_sql_query("SELECT state, population_lakh as population, literacy_rate FROM state_stats WHERE population_lakh<100 and literacy_rate>80 ",conn)
```

```python
#13. How many unique literacy rates are there in the dataset?
pd.read_sql_query("SELECT COUNT(DISTINCT(literacy_rate)) as unique_literacy_rates FROM state_stats",conn)
```

```python
pd.read_sql_query("SELECT state, avg(gdp_cr), avg(literacy_rate) FROM state_stats",conn)
```

```python
#14. Which states have both above-average GDP and above-average literacy rate?
pd.read_sql_query("SELECT state, gdp_cr, literacy_rate FROM state_stats WHERE gdp_cr > (SELECT avg(gdp_cr) FROM state_stats) and literacy_rate > (SELECT avg(literacy_rate) FROM state_stats)",conn)
```

```python
#15. List all states from the North-East region with literacy rate below 80.
pd.read_sql_query("SELECT region, state, literacy_rate FROM state_stats WHERE region='North-East' and literacy_rate<80 ",conn)
```
